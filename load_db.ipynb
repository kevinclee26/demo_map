```python
import pandas as pd
import random
import sqlalchemy
```

```python
import json
```

```python
with open('source_data/California_78_to_82.geojson') as f:
    cali_shape=json.load(f)
```

```python
cali_shape_list=[]
for each_feature in cali_shape['features']: 
    cali_shape_list.append(each_feature['geometry']['coordinates'])
```

```python
df=pd.DataFrame({'coordinates': cali_shape_list})
df.head()
```

```python
for idx, each_row in df.iterrows(): 
    df.loc[idx, 'value']=random.randint(0, 100)
```

```python
df
```

```python
engine=sqlalchemy.create_engine('sqlite:///data/db.sqlite')
```

```python
df['coordinates']=df['coordinates'].astype('str')
```

```python
df.to_sql('data', if_exists='replace', index=False, con=engine)
```
