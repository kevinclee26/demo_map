```python
import sqlalchemy as sql
```

```python
engine=sql.create_engine('sqlite:///data/db.sqlite')
```

```python
results=engine.execute('select * from data').all()
```

```python
cali_shape_list=[]
for each_result in results: 
    shape=eval(each_result[0])
    cali_shape_list.append(shape)
```
